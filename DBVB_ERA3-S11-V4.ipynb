```python
# create a mapping from characters to integers
stoi = { ch:i for i,ch in enumerate(chars) }
itos = { i:ch for i,ch in enumerate(chars) }
encode = lambda s: [stoi[c] for c in s] # encoder: take a string, output a list of integers
decode = lambda l: ''.join([itos[i] for i in l]) # decoder: take a list of integers, output a string
```

```python
def get_stats(ids):
    counts = {}
    for pair in zip(ids, ids[1:]): # Pythonic way to iterate consecutive elements
        counts[pair] = counts.get(pair, 0) + 1
    return counts

stats = get_stats(tokens)
```

```python
def merge(ids, pair, idx):
  # in the list of ints (ids), replace all consecutive occurences of pair with the new token idx
  newids = []
  i = 0
  while i < len(ids):
    # if we are not at the very last position AND the pair matches, replace it
    if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
      newids.append(idx)
      i += 2
    else:
      newids.append(ids[i])
      i += 1
  return newids
```

```python
# ---
vocab_size = 5000 # the desired final vocabulary size
num_merges = vocab_size - 256
ids = list(tokens) # copy so we don't destroy the original list

print ("Starting merging ... ")
merges = {} # (int, int) -> int
for i in range(num_merges):
  stats = get_stats(ids)
  pair = max(stats, key=stats.get)
  idx = 256 + i
  print(f"merging {pair} into a new token {idx}")
  ids = merge(ids, pair, idx)
  merges[pair] = idx

print ("Completed merging ... ")    
```

```python
print("Tokens length:", len(tokens))
print("Ids length:", len(ids))
print("Vocab size:", vocab_size)
print(f"Compression ratio: {len(tokens) / len(ids):.2f}X")
```

```python
vocab = {idx: bytes([idx]) for idx in range(256)}
for (p0, p1), idx in merges.items():
    vocab[idx] = vocab[p0] + vocab[p1]

def decode(ids):
  # given ids (list of integers), return Python string
  tokens = b"".join(vocab[idx] for idx in ids)
  text = tokens.decode("utf-8", errors="replace")
  return text

print(decode([128]))
```

```python
def encode(text):
  # given a string, return list of integers (the tokens)
  tokens = list(text.encode("utf-8"))
  while len(tokens) >= 2:
    stats = get_stats(tokens)
    pair = min(stats, key=lambda p: merges.get(p, float("inf")))
    if pair not in merges:
      break # nothing else can be merged
    idx = merges[pair]
    tokens = merge(tokens, pair, idx)
  return tokens
```

```python
import json
import torch  # Example using PyTorch
import base64

# Convert bytes to Base64-encoded string
def convert_bytes(obj):
    if isinstance(obj, bytes):
        return base64.b64encode(obj).decode('utf-8')  # Encode as Base64 string
    raise TypeError(f"Object of type {type(obj)} is not JSON serializable")

# Convert tensors to lists
def convert_tensors(obj):
    if isinstance(obj, torch.Tensor):  # Check if the object is a tensor
        return obj.tolist()
    raise TypeError(f"Object of type {type(obj)} is not JSON serializable")

# Specify the filename
jsonfilename = "encoder-kan.json"

# Serialize dictionary to JSON
with open(jsonfilename, 'w') as file:
    json.dump(vocab, file, default=convert_bytes, indent=4) # `indent` makes the JSON file more readable

print(f"Dictionary has been saved to {jsonfilename}")
```
